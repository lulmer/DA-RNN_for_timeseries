## Raven price : 

```python
r = requests.get('https://coinmarketcap.com/currencies/raven-protocol/historical-data/?start=20130429&end=20191116')
soup = BeautifulSoup(r.text, 'lxml')
```

```python
L1=[] 
table = soup.find('script', id='__NEXT_DATA__',type="application/json")
values=json.loads(table.text)
ones=values["props"]["initialState"]["cryptocurrency"]["ohlcvHistorical"]["4024"]["quotes"]
for i in range(len(ones)) :
    # [open high low close volume market_cap date]
    val=ones[i]["quote"]["USD"]
    L1.append([val["open"],val["high"],val["low"],val["close"],val["volume"],val["market_cap"],val["timestamp"]])
```

```python
import pandas as pd
raven = pd.DataFrame(L1, columns = ["open", "high", "low", "close", "volume","market_cap","Date"]) 
raven["Date"]=pd.to_datetime(raven["Date"])
```

```python
import matplotlib.pyplot as plt
plt.plot(raven["close"])
plt.title("Close price of raven")
plt.show()
```
